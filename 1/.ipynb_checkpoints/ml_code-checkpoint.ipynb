```python
#导入包
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Figures inline and set visualization style
%matplotlib inline
sns.set()
```

```python
#导入测试集和训练集
df_train=pd.read_csv("./train.csv")
df_test=pd.read_csv("./test.csv")
#查看训练集前四项
df_train.head(n=4)
```

```python
#查看各个变量类型和数目
df_train.info()
```

```python
#绘制存活人数频率直方图,进行第一次预测
sns.countplot(x="Survived",data=df_train)
#由图可以看出,死亡人数>存活人数
#所以可以假设,所有人都死亡
df_test["Survived"]=0
df_test[["PassengerId","Survived"]].to_csv('./data/predictions/no_survivors.csv',index=False)
#将no_survivors.csv提交到Kaggle,结果为:0.62679
```

```python
#第二次预测,猜测性别对存活的影响
#将性别作为因子,进行绘图查看
sns.factorplot(x="Survived",col="Sex",kind="count",data=df_train)
#从下图可以看到,女性的存活率比男性的存活率高,所以可以假设所有男性全部死亡,全部女性存活
df_test['Survived']=df_test.Sex=="female"
df_test['Survived']=df_test.Survived.apply(lambda x:int(x))
df_test.head()
df_test[["PassengerId","Survived"]].to_csv("./data/predictions/female_survived.csv",index=False)
#提交到Kaggle,结果为:0.76555
```

```python
#猜测Pclass对Survived的影响
sns.factorplot(x="Survived",col="Pclass",kind="count",data=df_train)
#下图可以看出Pclass=3对存活率影响大,所以可以将Pclass=3的元组设置为死亡
df_test.loc[df_test.Pclass==3,'Survived']=0
df_test[["PassengerId","Survived"]].to_csv("./data/predictions/pclass_survived.csv",index=False)   
#提交到Kaggle,结果为:0.75598
```

```python
#使用Embarked对survived预测
sns.factorplot(x="Survived",col="Embarked",kind="count",data=df_train)
df_test.loc[df_test.Embarked=="S","Survived"]=0
df_test[["PassengerId","Survived"]].to_csv("./data/predictions/embarked_survived.csv",index=False)
#提交到Kaggle,结果:0.72248
```

```python
#Fare对Survived影响
sns.distplot(df_train.Fare,kde=False)
```

```python
sns.stripplot(x="Survived",y="Fare",data=df_train,alpha=0.3,jitter=True)
```

```python
sns.swarmplot(x="Survived",y="Fare",data=df_train)
```

```python
sns.lmplot(x='Age', y='Fare', hue='Survived', data=df_train, fit_reg=False, scatter_kws={'alpha':0.5});
#从上图可以看出,票价高的存活率高点
```
